```python
##Useful when we have a large number of unique categories, and we want to use this variable as a feature in our ML model
##we replace each category in the categorical variable with a numerical value based on the mean or median of the target variable for the category.
```

```python
import pandas as pd

#Create a sample dataframe with a categorical variable and a target variable
df=pd.DataFrame({
    'city':['New York','London','Paris','Tokyo','New York','Paris'],
    'price':[200,150,300,250,180,320]
})
```

```python
df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df
```

```python
df[['price','city_encoded']]
```

```python
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head()
```

```python
mean_total_bill=tips.groupby('time')['total_bill'].mean().to_dict()
```

```python
mean_total_bill
```

```python
tips['encoded_time']=tips['time'].map(mean_total_bill)
```

```python
tips[['time','encoded_time']]
```
