```python
#As machine doesn't underesrtand significance or order of categorical data, categorical data is represented as Numerical Data
#This process of converting categorical data to numerical data for machine learning algorithms is called Data Encoding
#Like say, BE, ME, and PhD if we convert these 3 to numbers then ML algorithms can understand the significance of degrees
```

```python
#1. Nominal / One Hot Encoding
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
##Create a simple dataframe
df=pd.DataFrame({
    'color':['red','blue','green','green','red','blue']})
```

```python
df
```

```python
##Create an instance of OneHotEncoder
encoder=OneHotEncoder()
```

```python
## Perform fit and Transform
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
##for new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```

```python
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
encode=OneHotEncoder()
encoded_tips=pd.DataFrame(encode.fit_transform(tips[['sex','smoker','day','time']]).toarray(),columns=encode.get_feature_names_out())
```

```python
encoded_tips.head()
```

```python
final_tips_encoded=pd.concat([tips,encoded_tips],axis=1)
```

```python
final_tips_encoded.head()
```
