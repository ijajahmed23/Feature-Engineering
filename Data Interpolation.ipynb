```python
###Linear Interpolation
```

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
##interpolate the data using Linear Interpolation
x_new=np.linspace(1,5,10) ##create new 10 x values
y_inetrp=np.interp(x_new,x,y)
```

```python
y_inetrp
```

```python
plt.scatter(x_new,y_inetrp)
```

```python
###Cubic Interpolation with Spicy
```

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
##Creating a cubic interpolation function
f=interp1d(x,y,kind='cubic')
```

```python
#interpolate the data
x_new=np.linspace(1,5,20)
y_interp1=f(x_new)
```

```python
y_interp1
```

```python
plt.scatter(x_new,y_interp1)
```

```python
##Polynomial Interpolation
```

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
```

```python
#interpolate the data using polynomial ibnterpolation
p=np.polyfit(x,y,2)
```

```python
#interpolate the data
x_new1=np.linspace(1,5,20) ##create new X values
y_interp2=np.polyval(p,x_new1) ##interpolate y values
```

```python
y_interp2
```

```python
plt.scatter(x_new1,y_interp2)
```
